```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler, LabelBinarizer
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
iris = load_iris()
X = iris.data
Y = iris.target
target_names = iris.target_names
target_names
```

```python
#Normalising
scaler = StandardScaler()
X_norm = scaler.fit_transform(X)

#One-Hot Encoding
# encoder = LabelBinarizer()
# Y_encode = encoder.fit_transform(Y)

X_train, X_test, Y_train, Y_test = train_test_split(X_norm, Y, test_size=0.2, random_state=20)
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.losses import SparseCategoricalCrossentropy
```

```python
model = Sequential()
model.add(Dense(10, activation='relu'))
model.add(Dense(5, activation='relu'))
model.add(Dense(3, activation='softmax'))

# model.compile(loss=SparseCategoricalCrossentropy(from_logits = True))

# model.fit(X_train, Y_train, epochs=100)

# logits = model(X_train)

# f_x = tensorflow.nn.softmax(logits)

# model=Sequential(
#     Dense(10, activation="relu"),
#     Dense(5, activation="relu"),
#     Dense(3, activation="softmax")

model.compile(loss=SparseCategoricalCrossentropy(from_logits=False), metrics=['accuracy'])

model.fit(X_train, Y_train, epochs=100)
```

```python
loss, accuracy = model.evaluate(X_test, Y_test)
print(f"Test Accuracy: {accuracy * 100:.2f}%")
```

```python
import joblib
model.save("iris_nn_model.keras")
joblib.dump(scaler, "scaler.pkl")
```
